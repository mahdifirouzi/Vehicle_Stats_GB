```python
import pandas as pd
import os
# Importing the save_dataframe function from the data_saver module
from data_saver import save_dataframe
from filter_year import filter_by_year_quarter
from filter_range import filter_by_year_range
#from KNN_imputation import rkNN_imputation
```

```python
# Load the CSV file into a DataFrame from the current directory
file_name = "df_VEH0120_GB.csv"
current_directory = os.getcwd()

# Construct the full file path
file_path = os.path.join(current_directory, file_name)

df_Vehicle_Stat_UK_ALL = pd.read_csv(file_path)

# Filter the dataframe for "Cars" in the 'BodyType' column
Car_Stat_UK_All = df_Vehicle_Stat_UK_ALL[df_Vehicle_Stat_UK_ALL['BodyType'] == 'Cars']

# Filter the dataframe for "Battery electric" in the 'Fuel' column
BECar_Stat_UK_All = Car_Stat_UK_All[Car_Stat_UK_All['Fuel'] == 'Battery electric']

# Filter the dataframe for "Licensed" in the 'LicenceStatus' column
BECar_Stat_UK_Registered = BECar_Stat_UK_All[BECar_Stat_UK_All['LicenceStatus'] == 'Licensed']


BECar_Stat_UK_Registered.head()  # Show the first few rows of the final dataframe for inspection
```

```python
BECar_Stat_UK_Registered= BECar_Stat_UK_Registered.drop(columns=['BodyType', 'Fuel', 'LicenceStatus'])
```

```python
BECar_Stat_UK_Registered.head()  # Show the first few rows of the final dataframe for inspection
```

```python
# Assuming your DataFrame has a column 'Year_Quarter' like '2024Q2'
Year=2024
Quarter=2
BECar_Stat_UK_Registered_YQ = filter_by_year_quarter(BECar_Stat_UK_Registered, Year, Quarter, f'No.')  # For Q2 of 2024
BECar_Stat_UK_Registered_YQ.head()
BECar_Stat_UK_Registered_YQ = BECar_Stat_UK_Registered_YQ[BECar_Stat_UK_Registered_YQ['No.'] != 0]
```

```python
BECar_Stat_UK_Registered_2024_2020_Q1 = filter_by_year_range(BECar_Stat_UK_Registered, 2024, 2020, 1)  # For Q1, from 2024 to 2020
BECar_Stat_UK_Registered_2024_2020_Q1.head()
```

```python
save_dataframe(BECar_Stat_UK_Registered_YQ, f"BECar_Stat_UK_Registered_{Year}Q{Quarter}", file_format="csv", save_index=False)
#save_dataframe(BECar_Stat_UK_Registered_2024_2020_Q1, "BECar_Stat_UK_Registered_2024_2020_Q1", file_format="csv", save_index=False)
BECar_Stat_UK_Registered_YQ
```

```python
# Filter the dataset into two parts:
# 1. Rows where Model is 'MISSING'
df_missing = BECar_Stat_UK_Registered_YQ[(BECar_Stat_UK_Registered_YQ['Model'] == 'MISSING') | (BECar_Stat_UK_Registered_YQ['No.'] < 11)]

# 2. The rest of the rows
df_not_missing = BECar_Stat_UK_Registered_YQ[~((BECar_Stat_UK_Registered_YQ['Model'] == 'MISSING') | (BECar_Stat_UK_Registered_YQ['No.'] < 11))]
```

```python
df_not_missing
```

```python
df_missing
```

```python
# Group by 'Make' and find the vehicle with the highest registrations within each make
most_registered_per_make = df_not_missing.loc[df_not_missing.groupby('Make')['No.'].idxmax()]

# Find the vehicle with the highest registrations in the whole dataset
most_registered_overall = df_not_missing.loc[df_not_missing['No.'].idxmax()]

# Display results
most_registered_per_make
```

```python
df_missing
```

```python
df_missing_updated = df_missing.copy()

# Add columns to track previous values
df_missing_updated['Previous_Make'] = df_missing_updated['Make']
df_missing_updated['Previous_GenModel'] = df_missing_updated['GenModel']
df_missing_updated['Previous_Model'] = df_missing_updated['Model']

```

```python
df_missing_updated
```

```python
for idx, row in df_missing_updated.iterrows():
    if row['Make'] in most_registered_per_make['Make'].values:
        replacement = most_registered_per_make[most_registered_per_make['Make'] == row['Make']].iloc[0]
        df_missing_updated.at[idx, 'Make'] = replacement['Make']
        df_missing_updated.at[idx, 'GenModel'] = replacement['GenModel']
        df_missing_updated.at[idx, 'Model'] = replacement['Model']
    else:
        df_missing_updated.at[idx, 'Make'] = most_registered_overall['Make']
        df_missing_updated.at[idx, 'GenModel'] = most_registered_overall['GenModel']
        df_missing_updated.at[idx, 'Model'] = most_registered_overall['Model']
```

```python
df_missing_updated
```

```python
# Merge df_missing_updated with df_not_missing
merged_df = pd.concat([df_missing_updated, df_not_missing], ignore_index=True)

# Drop the columns for previous values
merged_df = merged_df.drop(columns=['Previous_Make', 'Previous_GenModel', 'Previous_Model'], errors='ignore')

# Display or save the result
print(merged_df.head())
merged_df
```

```python
# Group by 'Make', 'GenModel', and 'Model' and sum the 'No. BEC' column
merged_df = merged_df.groupby(['Make', 'GenModel', 'Model'], as_index=False).agg({'No.': 'sum'})
print(merged_df.head())
merged_df
```

```python
save_dataframe(merged_df, f"BECar_Stat_UK_Registered_{Year}Q{Quarter}_Cleaned", file_format="csv", save_index=False)
```
