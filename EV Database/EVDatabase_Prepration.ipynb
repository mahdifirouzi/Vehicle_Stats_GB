```python
#this file use two csv of scrapped data for create the UK EV sepecification file
import pandas as pd
import numpy as np
import os
```

```python
current_directory = os.getcwd()

file_name_ev_data_market = "ev_data_on_the_market.csv"
file_name_ev_data_second_hand = "ev_data_second_hand.csv"
file_name_comparison_file = "Market_Second_hand_Compare.csv"
file_name_ev_specs = "EV_Specifications.xlsx"
file_name_ev_specs_uk = "EV_Specifications_UK.csv"
file_name_ev_specs_nl = "EV_Specifications_NL.csv"
file_name_ev_specs_ge = "EV_Specifications_GE.csv"
file_name_Vauxhall  = "EV_Specifications_Vauxhall_metric.csv"
file_name_ev_specs_uk_full = "EV_Specifications_UK_full.csv"

file_path_ev_data_market = os.path.join(current_directory, file_name_ev_data_market)
file_path_ev_data_second_hand = os.path.join(current_directory, file_name_ev_data_second_hand)
file_path_comparison_file = os.path.join(current_directory, file_name_comparison_file)
file_path_ev_specs = os.path.join(current_directory, file_name_ev_specs)
file_path_ev_specs_uk = os.path.join(current_directory, file_name_ev_specs_uk)
file_path_ev_specs_nl = os.path.join(current_directory, file_name_ev_specs_nl)
file_path_ev_specs_ge = os.path.join(current_directory, file_name_ev_specs_ge)
file_path_Vauxhall = os.path.join(current_directory, file_name_Vauxhall)
file_path_ev_specs_uk_full = os.path.join(current_directory, file_name_ev_specs_uk_full)
```

```python
# Load the datasets
ev_data_market = pd.read_csv(file_path_ev_data_market)
ev_data_second_hand = pd.read_csv(file_path_ev_data_second_hand)
#comparison_file = pd.read_excel('/mnt/data/Market_Second_hand_Compare.xlsx', sheet_name=None)
```

```python
# We will merge the EV data from the market with the second-hand data and add the appropriate columns

# Add 'Market Status' to the second-hand dataset
ev_data_second_hand['Market Status'] = 'Second Hand'

# Add 'Market Status' to the market dataset
ev_data_market['Market Status'] = 'Current'

ev_data_market['Index EV'] = ev_data_market.index + 1
ev_data_second_hand['Index EV'] = ev_data_second_hand.index + 1
```

```python

# Merge the datasets
merged_data = pd.concat([ev_data_market, ev_data_second_hand])
merged_data = merged_data.reset_index(drop=True)
merged_data['Index EV CSH'] = merged_data.index + 1
# Move the 'Market Status' column to the 3rd position
merged_data.insert(2, 'Market Status', merged_data.pop('Market Status'))
merged_data.insert(0, 'Index EV CSH', merged_data.pop('Index EV CSH'))

# Save the resulting dataset to a CSV file
merged_data.to_csv(file_path_comparison_file, index=False)
merged_data
```

```python
#polish_EV_Spec_data_for Size of Battery Extraction
# Put the merged data to ev specs df
ev_specs = merged_data
```

```python
# Replace 'N/A' strings with actual NaN values for easier filtering
## Remove stars before the price in the UK price column (assuming the price is in a column named 'EV Price UK')
ev_specs['EV Price in the United Kingdom after incentives'] = ev_specs['EV Price in the United Kingdom after incentives'].replace(r'^\*', '', regex=True)
ev_specs['EV Price in The Netherlands before incentives'] = ev_specs['EV Price in The Netherlands before incentives'].replace(r'^\*', '', regex=True)
ev_specs['EV Price in Germany before incentives'] = ev_specs['EV Price in Germany before incentives'].replace(r'^\*', '', regex=True)
ev_specs
```

```python
# Convert 'NaN' strings to actual NaN
ev_specs['EV Price in the United Kingdom after incentives'] = ev_specs['EV Price in the United Kingdom after incentives'].replace('NaN', np.nan)

# Remove commas from the 'EV Price UK' column and convert it to numeric
ev_specs['EV Price in the United Kingdom after incentives'] = ev_specs['EV Price in the United Kingdom after incentives'].replace({',': ''}, regex=True).astype(float)

# Convert 'NaN' strings to actual NaN
ev_specs['EV Price in The Netherlands before incentives'] = ev_specs['EV Price in The Netherlands before incentives'].replace('NaN', np.nan)
# Remove commas from the 'EV Price UK' column and convert it to numeric
ev_specs['EV Price in The Netherlands before incentives'] = ev_specs['EV Price in The Netherlands before incentives'].replace({',': ''}, regex=True).astype(float)

# Convert 'NaN' strings to actual NaN
ev_specs['EV Price in Germany before incentives'] = ev_specs['EV Price in Germany before incentives'].replace('NaN', np.nan)
# Remove commas from the 'EV Price UK' column and convert it to numeric
ev_specs['EV Price in Germany before incentives'] = ev_specs['EV Price in Germany before incentives'].replace({',': ''}, regex=True).astype(float)

# Convert the column to numeric, forcing errors to NaN (invalid entries will become NaN)
ev_specs['EV Price in the United Kingdom after incentives'] = pd.to_numeric(ev_specs['EV Price in the United Kingdom after incentives'], errors='coerce')
ev_specs['EV Price in Germany before incentives'] = pd.to_numeric(ev_specs['EV Price in Germany before incentives'], errors='coerce')
ev_specs['EV Price in The Netherlands before incentives'] = pd.to_numeric(ev_specs['EV Price in The Netherlands before incentives'], errors='coerce')

# Filter the data where the column has numeric values and is not NaN
filtered_data_UK = ev_specs[ev_specs['EV Price in the United Kingdom after incentives'].notna()]
#filtered_data_UK = filtered_data_UK.drop(columns=["EV Price in The Netherlands before incentives","EV Price in Germany before incentives"])
#filtered_data_UK
filtered_data_NL = ev_specs[ev_specs['EV Price in The Netherlands before incentives'].notna()]
filtered_data_GE = ev_specs[ev_specs['EV Price in Germany before incentives'].notna()]
filtered_data_UK
```

```python
filtered_data_UK.to_csv(file_path_ev_specs_uk, index=False)
filtered_data_GE.to_csv(file_path_ev_specs_ge, index=False)
filtered_data_NL.to_csv(file_path_ev_specs_nl, index=False)
```

```python
# Create a 'unique' column where True means the name is not unique (appears more than once)
df['not-unique'] = df.duplicated(subset=['EV Full Name'], keep=False)
df = df.reset_index(drop=False)
df
```

```python
from datetime import datetime

def process_date(date_str):
    try:
        # Try to parse date in 'Sep-23' format
        return datetime.strptime(date_str, '%b-%y').strftime('%y/%m/01')
    except ValueError:
        # If not in 'Sep-23' format, check for date range like 'November 2020 - March 2021'
        try:
            start_date = date_str.split(' - ')[0]
            return datetime.strptime(start_date, '%B %Y').strftime('%y/%m/01')
        except ValueError:
            return None  # In case it doesn't match any expected format
```

```python
# Apply the processing function to the 'Date Available' column and create 'processed_dates' column shows appearance on the market
df['processed_dates'] = df['Date Available'].apply(process_date)
```

```python
df_last_market_release = df.loc[df.groupby('EV Full Name')['processed_dates'].idxmax()]
```
